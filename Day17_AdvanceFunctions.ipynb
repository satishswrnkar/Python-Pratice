### Scoping 

```python
%%writefile scope1.py

x = 5 # global variable

def func(x):
    x = x * 2
    print("Inside x is : ", x)


print("Before x is : ", x)
func(x)
print("After x is : ", x)
```

```python
%%writefile scope2.py

x = 10

def func():
    print("Inside Value of X is : ", x)
    print("Inside Value of x**2 is: ", x**2)

print("Before x is : ", x)
func()
print("After x is : ", x)
```

```python
!python scope2.py
```

##### Rules of Scope

1. You can Access global varaibles inside local scope
2. You can not Modify global varaibles inside local scope without permission

```python
%%writefile scope4.py

x = 5

def change():
    global x
    x = x ** 2 
    print("Inside Value of x is : ", x)

print("Before x is : ", x)
change()
print("After x is : ", x)
```

```python
!python scope4.py
```
